```python
from google.colab import drive
drive.mount('/content/drive')
!pwd
%cd '/content/drive/My Drive/Colab Notebooks/Masters Project/Challenge 1/Dictionary'
!pwd


import re
from collections import Counter
def load_text_file(file_path):
    with open(file_path, 'r') as file:
        return file.read()


# using this function below we will first of convert all the words from the text into the lowercase letters because most
#of the time words in dictonary are in lower case , so we can match and compare them in effective way
def get_words(text):
    words = re.findall(r'\w+', text.lower())
    return words

def get_edits(word):
    letters = 'abcdefghijklmnopqrstuvwxyz'


    splits = [(word[:i], word[i:]) for i in range(len(word) + 1)]
    inserts = [left + letter + right for left, right in splits for letter in letters]

    deletes = [left + right[1:] for left, right in splits if right]
    replaces = [left + letter + right[1:] for left, right in splits for letter in letters]
    transposes = [left + right[1] + right[0] + right[2:] for left, right in splits if len(right) > 1]
    return set(deletes + replaces + transposes + inserts)

mistakes_corrected_words= [] #this is a variable to store the mistakes with corresponding corrections
file_mistakes_corrected_words = open("mistakes_corrected_words.txt", "w") #open a file for mistakes and their corrections
#following is the defination of contain number
def contains_number(word):
    return any(char.isdigit() for char in word)

def correct_spelling_mistakes(text):
    words = get_words(text)
    word_counts = Counter(words)
    corrected_words = []

    num_corrected = 0
    #if not word.isnummeric()
    for word in words:
        #if not contains_number(input_string):
        if word not in dictionary and not contains_number(word):
            candidate_edits = get_edits(word)
            valid_edits = candidate_edits.intersection(dictionary)
            if valid_edits:
                corrected_word = max(valid_edits, key=lambda w: word_counts.get(w, 0))
                corrected_words.append(corrected_word)
                num_corrected += 1









                #print("#"+str(num_corrected)+"        ("+word+" ------> "+corrected_word+")")
                file_mistakes_corrected_words.writelines("#"+str(num_corrected)+":"+word+"--->"+corrected_word+"\n")








            else:
                corrected_words.append(word)
        else:
            corrected_words.append(word)

    return ' '.join(corrected_words), num_corrected, len(words)

# Path to the text file
text_file = 'text.txt'

# Path to the dictionary file
dictionary_file = 'dictionary.txt'

# Load the dictionary
dictionary = set(get_words(load_text_file(dictionary_file)))

# Load the text file
text = load_text_file(text_file)

# Correct spelling mistakes and get correction statistics
corrected_text, num_corrected, num_words = correct_spelling_mistakes(text)

# Print the corrected text
#print("Original text:")
#hprint(text)

#print("\nCorrected text:")
#print(corrected_text)

# Print correction statistics
print(f"\nNumber of corrected words: {num_corrected}")
print(f"Total number of words: {num_words}")

# Path to save the corrected text
fName=input("Enter your name: ")
output_file = fName+'.txt'

# Save the corrected text to a file
with open(output_file, 'w') as file:
    file.write(corrected_text)
print(f"\nCorrected text saved to '{output_file}'.")
file_mistakes_corrected_words.close()
```
